# Converting duration in minutes

```python
print(df_movies['duration'])

minutes = []

for i in range(len(df_movies['duration'])) : 

    duration_split = df_movies['duration'][i].split()

    if len(duration_split) == 4 :
        minutes.append(int(duration_split[0])*60 + int(duration_split[2]))
    else :
        minutes.append(int(duration_split[0]))
    

minutes
```

```python
def h_to_m(df) :
    minutes = []

    for i in range(len(df['duration'])) : 

        duration_split = df['duration'][i].split()

        if len(duration_split) == 4 :
            minutes.append(int(duration_split[0])*60 + int(duration_split[2]))
        else :
            minutes.append(int(duration_split[0]))

    df['duration_minutes'] = minutes
    return df['duration_minutes'] 
```

```python
h_to_m(df_movies)
df_movies
```

```python
h_to_m(df_series)
df_series
```
